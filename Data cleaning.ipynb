```python
import pandas as pd
df = pd.read_excel(r'E:\datasets\Customer Call List.xlsx')
df
```

```python
# 1 removing duplicates

df.drop_duplicates(subset=['CustomerID'], keep='last', inplace=True)
(df)
```

```python
df.drop(columns = 'Not_Useful_Column', inplace = True)
(df)
```

```python
# removing special chars from column
'''
to do that you will use the strip function from str class
this function can be strip (both right and left) / lstrip (left)/ rstrip (right)

also you can use replace fucntion
and so on 
that means that all string function you can use
'''
df['Last_Name'] = df['Last_Name'].str.strip('\/._12345 ')
(df)
```

```python
# removing NaN
df['Last_Name'] = df['Last_Name'].apply(lambda x:str(x))
df['Last_Name'] = df['Last_Name'].replace('nan','')
df = df[df['Last_Name'] != '']
(df)
```

```python
# clean Phone_Number

# 1 reformat it

# first thing remove all un acceptalble rows
df['Phone_Number'] = df["Phone_Number"].str.replace('-','')
df['Phone_Number'] = df["Phone_Number"].str.replace('|','')
df['Phone_Number'] = df["Phone_Number"].str.replace('/','')
(df['Phone_Number'])
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:str(x))
df['Phone_Number'] = df['Phone_Number'].replace('nan','')
df = df[df['Phone_Number'] != ('')]
df = df[df['Phone_Number'] != ('Na')]
# now reformat it
df['Phone_Number'] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
df
```

```python
# replacing Y and N


df['Paying Customer'] = df['Paying Customer'].apply(lambda x: x.replace('N', 'No')  if x == 'N' else x)
df['Paying Customer'] = df['Paying Customer'].apply(lambda x: x.replace('Y', 'Yes')  if x == 'Y' else x)

df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: x.replace('N', 'No')  if x == 'N' else x)
df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: x.replace('Y', 'Yes')  if x == 'Y' else x)
(df)
```

```python
for i in df.index:
    if df.loc[i, 'Do_Not_Contact'] == 'Yes':
        df.drop(i, inplace=True)
(df)
```

```python
df.set_index('CustomerID', inplace=True)
(df)
```

```python
df.to_excel(r'E:\datasets\Customer Call List After data clean.xlsx')
```
